```python
#Laboratorio 6

import csv
import copy

# Definición del diccionario de un vehículo
myVehicle = {
    "vin": "<empty>",
    "make": "<empty>",
    "model": "<empty>",
    "year": 0,
    "range": 0,
    "topSpeed": 0,
    "zeroSixty": 0.0,
    "mileage": 0
}

# Mostrar las claves y valores iniciales
for key, value in myVehicle.items():
    print(f"{key} : {value}")

# Lista vacía para el inventario
myInventoryList = []

# Leer el archivo CSV y crear inventario en memoria
with open('car_fleet.csv') as csvFile:
    csvReader = csv.reader(csvFile, delimiter=',')
    lineCount = 0
    for row in csvReader:
        if lineCount == 0:
            print(f'Column names are: {", ".join(row)}')
            lineCount += 1
        else:
            print(f'vin: {row[0]} make: {row[1]}, model: {row[2]}, year: {row[3]}, '
                  f'range: {row[4]}, topSpeed: {row[5]}, zeroSixty: {row[6]}, mileage: {row[7]}')
            
            # Crear copia profunda del diccionario base
            currentVehicle = copy.deepcopy(myVehicle)
            
            # Asignar valores del CSV al diccionario
            currentVehicle["vin"] = row[0]
            currentVehicle["make"] = row[1]
            currentVehicle["model"] = row[2]
            currentVehicle["year"] = row[3]
            currentVehicle["range"] = row[4]
            currentVehicle["topSpeed"] = row[5]
            currentVehicle["zeroSixty"] = row[6]
            currentVehicle["mileage"] = row[7]
            
            # Agregar el vehículo a la lista
            myInventoryList.append(currentVehicle)
            lineCount += 1

    print(f'Processed {lineCount} lines.')

# Mostrar inventario completo
for myCarProperties in myInventoryList:
    for key, value in myCarProperties.items():
        print(f"{key} : {value}")
        print("-----")


#Fin laboratorio 6
```

```python
#Laboratorio 8

# Importar módulo para generar números aleatorios
import random

# Mensaje de bienvenida
print("Welcome to Guess the Number!")
print("The rules are simple. I will think of a number, and you will try to guess it.")

# Generar un número aleatorio entre 1 y 10
number = random.randint(1, 10)

# Variable para controlar si el usuario adivinó correctamente
isGuessRight = False

# Bucle while: se repetirá hasta que el usuario adivine el número
while not isGuessRight:
    guess = input("Guess a number between 1 and 10: ")  # Pedir entrada al usuario
    if int(guess) == number:  # Comparar la entrada con el número generado
        print("You guessed {}. That is correct! You win!".format(guess))
        isGuessRight = True  # Salir del bucle
    else:
        print("You guessed {}. Sorry, that isn’t it. Try again.".format(guess))


#Fin laboratorio 8
```
